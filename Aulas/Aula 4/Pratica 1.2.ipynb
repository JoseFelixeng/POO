```python
class Pessoa:
    '''Representa uma pessoa a ser vacinada.'''

    def __init__(self, nome, idade, doses_vac, tipo_vac):
        '''utilizada para inserir os dados iniciais do paciente'''
        self.__nome = nome
        self.__idade = idade
        self.__doses_vac = doses_vac
        self.__tipo_vac = tipo_vac
        
    def __str__(self):
        '''utilizada para a saida das informações do paciente'''
        return (f'Nome: {self.__nome} - Idade: {self.__idade} ({self.__doses_vac} doses do tipo {self.__tipo_vac})')
    
    def toma_vacina(self, tipo_vac, doses_vac):
        '''utilizado para identificar o tipo correto e a quantidade de doses já tomadas'''
        if tipo_vac == 'A' and doses_vac == 0:
            print('Primeira dose Tipo A')
            self.__doses_vac += 1
        elif tipo_vac == 'A'  and  doses_vac == 1:
            print('Segunda dose Tipo A')
            self.__doses_vac += 1    
        elif tipo_vac == '':
            print('O Paciente tomou outro tipo de vacina')
        
        if tipo_vac == 'B' and doses_vac == 0:
            print('Primeira dose Tipo B')
            self.__doses_vac += 1
        elif tipo_vac == 'B'  and  doses_vac == 1:
            print('Segunda dose Tipo B')
            self.__doses_vac += 1
        elif tipo_vac == '':
            print('O Paciente tomou outro tipo de vacina')

        if tipo_vac == 'C' and doses_vac == 0:
            print('Primeira dose Tipo C')
            self.__doses_vac += 1
        elif tipo_vac == 'C'  and  doses_vac == 1:
            print('Segunda dose Tipo C')
            self.__doses_vac += 1
        elif tipo_vac == '':
            print('O Paciente tomou outro tipo de vacina')
    
    @property
    def nome(self):
        '''Exibi o nome do paciente'''
        return self.__nome
        
    @property
    def idade(self):
        '''Exibi a idade do paciente'''
        return self.__idade
    
    @property
    def doses_vac(self):
        '''retorna em qual dose o paciente está'''
        return self.__doses_vac
      
    @property
    def tipo_vac(self):
        '''Exibi o nome do paciente'''
        return self.tipo_vac
    
    @nome.setter
    def nome(self, novo_nome):
        '''atualiza o nome do paciente'''
        self.__nome = novo_nome
    
    @idade.setter
    def idade(self, n_idade):
        '''atualiza a idade do paciente'''
        self.__idade = n_idade
```
